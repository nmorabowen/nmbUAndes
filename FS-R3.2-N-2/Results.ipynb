```python
# Import dependency
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

from plotApeConfig import set_default_plot_params, blueAPE
set_default_plot_params()
from baseUnits import mm, cm, m, kgf, kN, tf, Pa, MPa, kip, ksi, N
```

```python
class results:
    def __init__(self, fileName, sheetName, parameters, drift=False):
        self.fileName=fileName
        self.sheetName=sheetName
        self._drift=drift
        self.parameters=parameters
        
        self.df=self.loadValues()
        self.driftCoeficient=self.driftCoefCalculation()
       
    def loadValues(self):
        header=0
        df=pd.read_excel(self.fileName,header=header,sheet_name=self.sheetName)
        return df
    
    def driftCoefCalculation(self):
        if self._drift is False:
            driftCoeficient=1
        else:
            driftCoeficient=100/(915*mm)
        return driftCoeficient        
    
    
    def plotResults(self, ax=None, color=blueAPE, linestyle='-',linewidth=1):
        df=self.df
        verticalUnits=self.parameters['verticalUnits']
        horizontalUnits=self.parameters['horizontalUnits']
        
        if ax is None:
            fig, ax = plt.subplots()
            
        disp=df['Displacement']
        react=df['Reaction']
        
        ax.plot(disp*self.driftCoeficient/(horizontalUnits), -react/(verticalUnits), color=color, linestyle=linestyle, linewidth=linewidth)
        
        pass
```

```python
parameters={
    "verticalUnits":kip,
    "horizontalUnits":mm,
    "Longitud":915*mm
}
```
